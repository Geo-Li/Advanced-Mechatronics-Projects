## Define functions for signal processing and data plotting

```python
def plot(X, Y, title, X_label, Y_label, figsize = (15, 8), comparison = None):
    plt.figure(figsize=figsize)
    plt.plot(X, Y, color = "black", label = "raw")
    if len(comparison) > 0:
        plt.plot(X, comparison, color = "red", label = "filtered")
    plt.title(title)
    plt.xlabel(X_label)
    plt.ylabel(Y_label)
    plt.show()
```

```python
# Function to perform FFT
def FFT_signals(time, signal):
    sample_rate = 1.0 / (time[1] - time[0])  # Sampling frequency
    num_samples = len(signal)  # Length of the signal
    freqs = np.fft.fftfreq(num_samples, d = 1/sample_rate)  # Frequency range
    fft_values = np.fft.fft(signal) / num_samples  # FFT and normalization
    
    # Only take the positive half of the spectrum
    positive_freqs = freqs[:num_samples//2]
    positive_fft_values = np.abs(fft_values[:num_samples//2])
    
    return positive_freqs, positive_fft_values
```

```python
def smooth_signals(signal, window_size):
    return signal.rolling(window=window_size, min_periods=window_size//2, center=True).mean()
```

```python
# Function to apply IIR low-pass filter
def low_pass_filter(signal, A, B):
    filtered_signal = np.zeros_like(signal)
    filtered_signal[0] = signal[0]
    for i in range(1, len(signal)):
        filtered_signal[i] = A * filtered_signal[i-1] + B * signal[i]
    return filtered_signal
```

```python
# Function to apply coeffiecient generated from https://fiiir.com/
# to signals data
def weighted_low_pass(signal, coefficient):
    # Create a cyclic iterator of coefficients
    cyclic_coefficient = cycle(coefficient)
    # Apply the cyclic coefficients to the signal
    return [next(cyclic_coefficient) * sig for sig in signal]
```

```python
def compare_signals(time, raw_signal, filtered_signal, filter_type):
    plot(X = time,
         Y = raw_signal,
         title = f"Comparison of raw signals and {filter_type} signals",
         X_label = "Time",
         Y_label = "Signal",
         figsize = (12, 5),
         comparison = filtered_signal)
```

```python
def compare_FFT(freqs, raw_fft_values, filtered_fft_values):
    plt.figure(figsize=(15, 8))
    plt.loglog(freqs, raw_fft_values, color = "black", label = "raw")
    plt.loglog(freqs, filtered_fft_values, color = "red", label = "filtered")
    plt.title(f"Comparison of raw signals' FFT and filtered signals' FFT")
    plt.xlabel("Freqency")
    plt.ylabel("Magnitude")
    plt.show()
```

```python
def apply_filters(signal, window_size, weight, coefficient):
    time = signal.iloc[:, 0]
    raw_signal = signal.iloc[:, 1]
    raw_positive_freqs, raw_positive_fft_values = FFT_signals(time, raw_signal)
    # Calculate and plot smoothed data against raw
    signal["Smoothed"] = smooth_signals(signal.iloc[:, 1], window_size)
    compare_signals(time = time,
                    raw_signal = raw_signal,
                    filtered_signal = signal["Smoothed"],
                    filter_type = "smoothed")
    # Calculate and plot smoothed data against raw by their FFTs
    smoothed_positive_freqs, smoothed_positive_fft_values = FFT_signals(time, signal["Smoothed"])
    compare_FFT(freqs = raw_positive_freqs,
                raw_fft_values = raw_positive_fft_values,
                filtered_fft_values = smoothed_positive_fft_values)
    # Calculate and plot low-pass filtered data against raw
    A = weight[0]
    B = weight[1]
    signal["Low-pass"] = low_pass_filter(signal.iloc[:, 1], A, B)
    compare_signals(time = time,
                    raw_signal = raw_signal,
                    filtered_signal = signal["Low-pass"],
                    filter_type = "low-pass")
    # Calculate and plot low-pass filtered data against raw by their FFTs
    low_pass_positive_freqs, low_pass_positive_fft_values = FFT_signals(time, signal["Low-pass"])
    compare_FFT(freqs = raw_positive_freqs,
                raw_fft_values = raw_positive_fft_values,
                filtered_fft_values = low_pass_positive_fft_values)
    # Calculate and plot low-pass (windowed-sinc FIR) data against raw
    signal["Weighted"] = weighted_low_pass(raw_signal, coefficient)
    compare_signals(time = time,
                    raw_signal = raw_signal,
                    filtered_signal = signal["Weighted"],
                    filter_type = "low-pass (windowed-sinc FIR)")
    # Calculate and plot smoothed data against raw by their FFTs
    low_pass_sinc_positive_freqs, low_pass_sinc_positive_fft_values = FFT_signals(time, signal["Weighted"])
    compare_FFT(freqs = raw_positive_freqs,
                raw_fft_values = raw_positive_fft_values,
                filtered_fft_values = low_pass_sinc_positive_fft_values)
    
```

## Plot signals with full designed FIR

```python
window_sizes = [1000, 330, 400, 96]
weights = [[0.95, 0.05], [0.95, 0.05], [0.9, 0.1], [0.95, 0.05]]
# Open and read the JSON file for low-pass (windowed-sinc FIR) data
with open("data/low_pass_sinc.json", 'r') as low_pass_sinc_file:
    low_pass_sinc = json.load(low_pass_sinc_file)
char_A_index = ord("A")
for index in range(len(signals)):
    print()
    print(f"To smooth signal {chr(char_A_index + index)}, I used window size: {window_sizes[index]}")
    print(f"To low-pass filter signal {chr(char_A_index + index)}, I set A as: {weights[index][0]} and B as: {weights[index][1]}")
    weighted_low_pass_info = low_pass_sinc[f"sig{chr(char_A_index + index)}"].items()
    info_str = f"To filter signal {chr(char_A_index + index)} with weights, I decided on "
    for key, value in weighted_low_pass_info:
        info_str += key + " to be: " + str(value) + ", "
    print(info_str[:-2])
print()
```
